## Dark matter density PDF evolving with redshift

```python
import os
import numpy as np
from matplotlib import pyplot as plt

from pathlib import Path
from abacusnbody.data.compaso_halo_catalog import CompaSOHaloCatalog
from abacusnbody.data import read_abacus
import astropy.table
from mockfactory import Catalog
from density import Density

plt.style.use(os.path.join(os.path.abspath('./'), 'plot_style.mplstyle'))
plots_dir = '/global/cfs/cdirs/desi/users/mpinon/density/plots/'
```

```python
# Y1 blinded data
data_fn = '/global/cfs/cdirs/desi/cosmosim/Abacus/AbacusSummit_base_c000_ph000/halos/z{:.3f}'

rlist = [6, 8, 10, 12]
densitylist = [np.load('/global/cfs/cdirs/desi/users/mpinon/density/AbacusSummit_base_c000_ph000_z1.100_cellsize{}_resamplertsc_delta_R.npy'.format(r)) for r in rlist]

cmap = plt.cm.coolwarm

plt.figure(figsize=(4, 3))
           
for i, r in enumerate(rlist[::-1]):
    vals, bins = np.histogram(densitylist[::-1][i], density=True, bins=400)
    plt.plot((bins[1:]+bins[:-1])/2, vals, color=cmap(i/3), label=r'$R={}$ Mpc/$h$'.format(r))
plt.xlim(-1, 3)
plt.xlabel(r'$\delta_R$')
plt.legend()
plt.savefig(os.path.join(plots_dir, 'abacus_smoothingscale_density_pdf.png'))
plt.show()
```

```python
zlist = [0.200, 0.500, 0.800, 1.400, 2.000]
densitylist = [np.load('/global/cfs/cdirs/desi/users/mpinon/density/AbacusSummit_base_c000_ph000_z{:.3f}'.format(z)+'_allp_cellsize10_resamplertsc_delta_R.npy') for z in zlist]

cmap = plt.cm.coolwarm

plt.figure(figsize=(4, 3))
           
for i, z in enumerate(zlist[::-1]):
    vals, bins = np.histogram(densitylist[::-1][i], density=True, bins=400)
    plt.plot((bins[1:]+bins[:-1])/2, vals, color=cmap(i/4), label=r'$z={}$'.format(z))
plt.text(1.62, 0.45, r'$R=10$ Mpc/$h$', fontsize=12, horizontalalignment='left')
plt.xlim(-1, 3)
plt.xlabel(r'$\delta_R$')
plt.legend()
#plt.title(r'dark matter smoothed density contrast $\delta_R$')
plt.savefig(os.path.join(plots_dir, 'abacus_evolving_density_pdf.png'))
plt.show()
```
