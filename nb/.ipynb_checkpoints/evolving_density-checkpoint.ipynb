## Dark matter density PDF evolving with redshift

```python
import sys
import os
import numpy as np
from matplotlib import pyplot as plt

from pathlib import Path
from abacusnbody.data.compaso_halo_catalog import CompaSOHaloCatalog
from abacusnbody.data import read_abacus
import astropy.table
from mockfactory import Catalog

sys.path.append('/global/u2/m/mpinon/density')
import density

plt.style.use(os.path.join(os.path.abspath('../'), 'plot_style.mplstyle'))
plots_dir = '/global/cfs/cdirs/desi/users/mpinon/density/plots/'
```

```python
from matplotlib.colors import LinearSegmentedColormap
base_colors = ['red', 'cornflowerblue']
```

```python
from scipy.interpolate import interp1d

def smooth_hist(x, y, xnew):
    return interp1d(x, y, 3)(xnew)
```

```python
# Y1 blinded data
data_fn = '/global/cfs/cdirs/desi/cosmosim/Abacus/AbacusSummit_base_c000_ph000/halos/z{:.3f}'

nbar = 9906847883/2000**3

xnew = np.linspace(-1, 3, 401)

rlist = [8, 10, 12, 16, 20]
cmap = LinearSegmentedColormap.from_list("mycmap", base_colors, N=len(rlist))
#densitylist = [np.load('/global/cfs/cdirs/desi/users/mpinon/density/AbacusSummit_base_c000_ph000_z1.100_cellsize{}_resamplertsc_delta_R.npy'.format(r)) for r in rlist]
densitylist = [np.load('/pscratch/sd/m/mpinon/densitysplits/AbacusSummit_2Gpc_z0.800_ph000_particles_nbar1.2384_cellsize20_resamplertophat_smoothingR{:02d}_delta_R_jax.npy'.format(r)) for r in rlist]
norms = np.array([(nbar * 4/3 * np.pi * r**3) for r in rlist])/np.array([200, 400, 600, 1000, 1000])
x = np.arange(0, 400)/norms[:, None] - 1 - 1/norms[:, None]/2

plt.figure(figsize=(3.2, 2.8))           
for i, r in enumerate(rlist[::-1]):
    vals, bins = np.histogram(densitylist[::-1][i], density=True, bins=x[::-1][i])
    newvals = smooth_hist((bins[1:]+bins[:-1])/2, vals, xnew)
    plt.plot(xnew, newvals, color=cmap(i/3), label=r'${}$'.format(r))
#plt.text(1.42, 0.4, r'$z = 1.1$', fontsize=12, horizontalalignment='left')
plt.xlim(-1, 3)
plt.xlabel(r'$\delta_R$')
plt.ylabel(r'$\mathcal{P}(\delta_R)$')
plt.legend(title=r'$R~[h^{-1}\mathrm{Mpc}]$')
plt.savefig(os.path.join(plots_dir, 'abacus_smoothingscale_density_pdf.pdf'), dpi=500)
plt.show()
```
